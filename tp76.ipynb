```python
import matplotlib.pyplot as plt
file = open("111.txt", mode = "r", encoding = "utf-8")
x = []
y = []
suma = []
for line in file:
    a = line.split(",")
    if (a[0] == "Средняя пенсия" and a[1] == "Забайкальский край" and a[2][:4] == "2018"):
        x.append(int(a[2][5:-3]))
        y.append(int(a[3]))
print(sum(y)/len(y))
plt.plot(x, y)
plt.title("Средняя пенсия за 18 год")
plt.xticks(x)
plt.show()
```
