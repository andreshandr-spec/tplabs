```python
import requests

class Myfile:
    def __init__(self, name, func):
        if func not in ["read", "write", "append", "url"]:
            print("error, invalid func")
        if name == '':
            print("error, name empty")
        self.name = name
        self.flag = func
    
    def read(self):
        if self.func == "read":
            with open(self.name, mode = 'r') as file:
                for _ in file.readlines():
                    a = open(self.name, mode = 'r').read
                    return a
        else:
            print("err: func not read")
    
    def write(self, string):
        if self.func == "write":
            open(self.name, mode = 'w').write(string)
        elif self.func == "append":
            open(self.name, mode = 'a').write(string)
        else:
            print("err: func not write or append")

    def read_url(self):
        if self.func == "url":
            try:
                t = requests.get(self.name)
                return t.text
            except requests.ConnectionError:
                return "err: site not found"
        else:
            print("err: func not url")
    
    def count_urls(self):
        if self.func == "url":
            try:
                t = requests.get(self.name)
                text = t.text
                count = text.count('href=')
                return count
            except requests.ConnectionError:
                return "err: site not found"
        else:
            print("err: func not url")

def write_url(self, name):
        if self.flag == "url":
            try:
                file = open(name, mode = 'w')
                t = requests.get(self.name)
                file.write(t.text)
            except requests.ConnectionError:
                print("err: site not found")
        else:
            print("err: func not url")
```
