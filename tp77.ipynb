```python
import requests

class Myfile:
    def __init__(self, name, flag):
        if flag not in ["read", "write", "append", "url"]:
            print("error, invalid flag")
        if name == '':
            print("error, name empty")
        self.name = name
        self.flag = flag
    
    def read(self):
        if self.flag == "read":
            a = open(self.name, mode = 'r').read
            return a
        else:
            print("err: flag not read")
    
    def write(self, string):
        if self.flag == "write":
            f = open(self.name, mode = 'w').write(string)
        elif self.flag == "append":
            f = open(self.name, mode = 'a').write(string)
        else:
            print("err: flag not write or append")
        f.close

    def read_url(self):
        if self.flag == "url":
            try:
                t = requests.get(self.name)
                return t.text
            except requests.ConnectionError:
                return "err: site not found"
        else:
            print("err: flag not url")
    
    def count_urls(self):
        if self.flag == "url":
            try:
                t = requests.get(self.name)
                text = t.text
                count = text.count('href=')
                return count
            except requests.ConnectionError:
                return "err: site not found"
        else:
            print("err: flag not url")

    def write_url(self, name):
        if self.flag == "url":
            try:
                file = open(name, mode = 'w')
                t = requests.get(self.name)
                file.write(t.text)
                file.close
            except requests.ConnectionError:
                print("err: site not found")
        else:
            print("err: flag not url")
```
