```python
file = open("63_lab.txt", mode = "r", encoding = "cp1251")
collection = []
s = []
chek = [0, 0, 0]
for line in file:
    a = line.split('\t')
    b = a[3].split(',')
    if b != ['']:
        for i in b:
            if i not in collection:
                collection.append(i)
            else:
                s.append(i)
for j in [0, 1, 2]:
    cnt = 0
    for i in collection:
        if s.count(i) > cnt and i not in chek:
            cnt = s.count(i)
            chek[j] = i
print(chek)
```
